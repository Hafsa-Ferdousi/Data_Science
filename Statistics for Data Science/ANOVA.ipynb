```python
import numpy as np
from scipy import stats

group_A = [70, 75, 80]
group_B = [60, 65, 70]
group_C = [85, 90, 95]

# ANOVA test
f_value, p_value = stats.f_oneway(group_A, group_B, group_C)

print("F-value:", f_value)
print("P-value:", p_value)
```

Since p-value (0.0025) < 0.05, we reject the null hypothesis and conclude that there is a significant difference among the group means.
