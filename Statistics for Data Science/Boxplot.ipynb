```python
import matplotlib.pyplot as plt

data = [30,40,50,60,70,100]
plt.boxplot(data)
plt.title("Basic Boxplot Example")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt

data1 = [30,40,50,70,60]
data2 = [0,25, 50 ,75, 100]

plt.boxplot([data1,data2], label=["dataSet1","Dataset2"])
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt

data= [10,12,14,15,16,18,20,100]

plt.boxplot(data,showfliers=True)
plt.title("Boxplot with Outliers")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.DataFrame({
    "Marks":[30,40,50,60,70,100]
                   })

data.boxplot(column="Marks")
plt.title("Boxplot using Pandas")
plt.show()
```
