```python
import matplotlib.pyplot as plt


x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 6]

plt.figure()
plt.scatter(x, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Simple Scatter Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

hours = [1, 2, 3, 4, 5, 6, 7]
marks = [40, 45, 50, 55, 65, 70, 80]

plt.figure()
plt.scatter(hours, marks)
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Hours": [1,2,3,4,5],
    "Marks": [35,45,55,60,70]
})

plt.figure()
plt.scatter(df["Hours"], df["Marks"])
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.title("Scatter Plot using DataFrame")
plt.show()
```
