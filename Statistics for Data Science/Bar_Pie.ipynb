```python
import matplotlib.pyplot as plt
import pandas as pd


outlook = [
    "sunny", "sunny", "overcast", "rainy", "rainy",
    "rainy", "overcast", "sunny", "sunny", "rainy",
    "overcast", "overcast"
]


counts = pd.Series(outlook).value_counts()


plt.figure()
plt.pie(
    counts.values,
    labels=counts.index,
    autopct="%1.1f%%"
)
plt.title("Outlook Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd


outlook = ["sunny","sunny","overcast","rainy","rainy","rainy","overcast","sunny","sunny","rainy"]


counts = pd.Series(outlook).value_counts()


plt.figure()
plt.bar(counts.index, counts.values)
plt.xlabel("Outlook")
plt.ylabel("Frequency")
plt.title("Bar Chart of Outlook")
plt.show()
```
