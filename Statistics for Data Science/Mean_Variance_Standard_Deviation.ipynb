```python
import numpy as np
import matplotlib.pyplot as plt

# Datasets
data1 = np.array([30, 40, 50, 60, 70])
data2 = np.array([0, 25, 50, 75, 100])

# Mean
mean1 = np.mean(data1)
mean2 = np.mean(data2)

# Variance
var1 = np.var(data1)
var2 = np.var(data2)

# Standard Deviation
std1 = np.std(data1)
std2 = np.std(data2)

print("Dataset 1")
print("Mean:", mean1)
print("Variance:", var1)
print("Standard Deviation:", std1)

print("\nDataset 2")
print("Mean:", mean2)
print("Variance:", var2)
print("Standard Deviation:", std2)

# Visualization
plt.figure()
plt.plot(data1, marker='o', label='Dataset 1')
plt.plot(data2, marker='s', label='Dataset 2')
plt.axhline(50, linestyle='--', label='Mean = 50')
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Dispersion Comparison of Two Datasets")
plt.legend()
plt.show()
```
