## 1) Python Data Structures

### Tuples

```python

# Tuples are immutable and support unpacking

point = (12.5, 7.0)
x, y = point
print("x:", x, "y:", y)

def arith(a,b):

  return a*b,a+b,a-b,[b,a]

mult,sum,subt,rev = arith(x,y)

print("x*y:", mult, "x+y:", sum,"x-y",subt,"rev",rev)
```

### Sets

```python

# Sets keep unique items; order is not guaranteed
a = {1,2,3,3,2}
b = {3,4,5}
print("a:", a)
print("Union:", a | b)
print("Intersection:", a & b)
print("Difference a-b:", a - b)

c = set()

c.add(7)
c.add(8)
print(c)
c.remove(7)
print(c)

if 5 in b:
  print("5 is in b")
else:
  print("5 is not in b")
```

### Dictionaries

```python
profile = {"id": 101, "name": "Ibrahim", "skills": ["Python","Pandas"]}
profile["gpa"] = 3.95      # add
profile["skills"].append("NumPy")  # mutate nested list
print("Keys:", list(profile.keys()))

for k, v in profile.items():
    print(k, "->", v)

student_db = {}

student_db[profile["id"]] = profile
print(student_db)
```

## 2) Python OOP — Classes & Objects

```python
class Counter:
    def __init__(self, start=0):
        self.value = int(start)
    def inc(self, step=1):
        self.value += step
    def dec(self,step=1):
        self.value -= step
    def reset(self):
        self.value = 0
    def __repr__(self):
        return f"Counter(value={self.value})"

c = Counter(5)
c.inc()
print(c)
c.inc(3)
print(c)
c.dec()
print(c)
c.dec(2)
print(c)
```

## 3) Advanced Concepts

### Modules

```python

# Create a small module file and import it
module_src = '''
def mean(xs):
    return sum(xs)/len(xs)

def mode(xs):
    from collections import Counter
    return Counter(xs).most_common(1)[0][0]
'''
with open('stats_helpers.py','w',encoding='utf-8') as f:
    f.write(module_src)

import stats_helpers as sh

print(sh.mean([1,2,3,4]), sh.mode([1,2,2,3]))
```

### Try…Except

```python

def safe_div(a,b):
    try:
        return a/b
    except ZeroDivisionError:
        return float("inf")  # convention for division by zero
    except TypeError as e:
        print("Type error:", e)
        return None

print(safe_div(10,2))
print(safe_div(1,0))
print(safe_div("x",5))
```
