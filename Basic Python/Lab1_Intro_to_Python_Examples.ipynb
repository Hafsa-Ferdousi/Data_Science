### Python Casting

```python

# Casting (type conversion)
x = "42"
y = int(x)       # string -> int
z = float(x)     # string -> float
s = str(3.5)     # number -> string
print(x, type(x))
print(y, type(y))
print(z, type(z))
print(s, type(s))
```

### Strings

```python

title = "statistics for data science"
print(title.title())
print(title.upper())
print(title)
print(title.replace("data", "information"))
print("Length:", len(title))
print("science starts at",title.find("science"))

# f-strings for formatting
student = "Zishan"
score = 93.25
print(f"{student} scored {score:.5f}% in Lab 1.")
```

### Booleans

```python

a = 10
b = 3
print("a > b:", a > b)
print("a == b:", a == b)
print("isinstance(a, int):", isinstance(a, int))
print("Compound:", (a > 50) and (b < 5))
```

## 2. Python Data Structures — Lists

```python
# Lists
grades = [78, 85, 92, 88]
print("Original:", grades)

# Indexing and slicing
print("First:", grades[0])
print("Slice 1..3:", grades[1:3])

# Mutation
grades[1] = 90
print("After mutation:", grades)

# Append & extend
grades.append(95)
print("After append:", grades)
grades.extend([67, 88])
print("After extend:", grades)
```

```python
demo_list = [5,9,"student",[5.8,'w',96]]

print(demo_list)
print(len(demo_list))

print(demo_list[3][2])
```

### Basic List Operations

```python

grades = [78, 90, 92, 88, 95, 67, 88]
print("Count of 88:", grades.count(88))
print("Index of 92:", grades.index(92))
print("Min, Max:", min(grades), max(grades))
print("Sum, Len, Mean:", sum(grades), len(grades), sum(grades)/len(grades))

# Sorting (returns None, sorts in place)
grades.sort()
print("Sorted:", grades)

# Sorted copy (does not mutate)
g2 = sorted(grades, reverse=True)
print("Reverse-sorted copy:", g2)
```

## 3. Python Control Flow

### If…Else

```python

mark = 88
if mark >= 80:
    grade = "A+"
    print("you got A+")
elif mark >= 75:
    grade = "A"
elif mark >= 70:
    grade = "A-"
else:
    grade = "C or below"
print("Mark:", mark, "Grade:", grade)
```

### While Loops

```python

# Sum natural numbers up to n using while
n = 5
total = 0
i = 1
while i <= n:
    total += i
    i += 1
print(total)
```

```python
x = 1
y = 1

for i in range(10):
  print(x)
  x,y = y,x+y
```

### For Loops & `range`

```python

# Iterate over a list
names = ["Saiful", "Nafis", "Sudipto"]
for person in names:
    print("Hello,", person)

# Iterate with range
for i in range(1, 6):  # 1..5
    print("Square of", i, "is", i*i)
```

## 4. Python Functions

```python
def mean(values):
    return sum(values)/len(values)

def letter_grade(mark):
    if mark >= 80:
        return "A"
    elif mark >= 70:
        return "B"
    elif mark >= 60:
        return "C"
    else:
        return "D/F"

scores = [72, 88, 91, 67, 83]
print("Mean:", mean(scores))
print("Grades:", [letter_grade(s) for s in scores])
```

```python
def fibonacci(n):
  x = 1
  y = 1

  for i in range(n-1):
    x,y = y,x+y

  return x
```

```python
def mult(x,y):
  return x*y
```
